```python
import pandas as pd
```

```python
df = pd.read_csv('data/art_data_loaded.csv')
```

```python
df.head()
```

```python
df.groupby(['type','style']).count()['index'].to_csv("temp.csv")
```

```python
df.groupby('style').count()
```

```python
df = df.groupby('style').sample(100,random_state=1)
```

```python
fine_grade_groups = dict()

indexes_to_groups = dict()


for x,y in enumerate(df.groupby('style')):
    fine_grade_groups[x] = [[],[]]
    

for group_ind,(name,group) in enumerate(df.groupby(['style'])):

    for ind,row in group.iterrows():
        indexes_to_groups[row['index']] = group_ind

    pairs = group['index'].to_list()

    cstyle = group.iloc[0]['style']

    neg_paris = df.loc[ (df['style'] !=cstyle) ]['index'].to_list()

    fine_grade_groups[group_ind][0].extend(pairs)
    fine_grade_groups[group_ind][1].extend(neg_paris)
```

```python
import json
```

```python
with open("style_sample_indexes_to_groups.json",'w')as f:
    f.write(json.dumps(indexes_to_groups))

with open("style_sample_pairs_groups.json",'w')as f:
    f.write(json.dumps(fine_grade_groups))
```
