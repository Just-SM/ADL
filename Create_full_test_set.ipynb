Creating the indexes of positive and negative pairs for images.

```python
fine_grade_groups = dict()

indexes_to_groups = dict()


for x,y in enumerate(df.groupby(['type','style','school'])):
    fine_grade_groups[x] = [[],[]]
    

for group_ind,(name,group) in enumerate(df.groupby(['type','style','school'])):

    for ind,row in group.iterrows():
        indexes_to_groups[row['index']] = group_ind

    pairs = group['index'].to_list()

    cschool = group.iloc[0]['school']
    ctype  = group.iloc[0]['type']
    cstyle = group.iloc[0]['style']

    neg_paris = df.loc[ (df['type'] != ctype) & (df['school'] != cschool) & (df['style'] !=cstyle) ]['index'].to_list()

    fine_grade_groups[group_ind][0].extend(pairs)
    fine_grade_groups[group_ind][1].extend(neg_paris)
```

```python
with open("indexes_to_groups.json",'w')as f:
    f.write(json.dumps(indexes_to_groups))

with open("pairs_groups.json",'w')as f:
    f.write(json.dumps(fine_grade_groups))
```
